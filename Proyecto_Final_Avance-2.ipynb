### Pasos de la Metodologa

#### 1 - Entendimiento del negocio / Problema

* Identificacin de los objetivos y necesidades del proyecto.
* Tener una formulacin inicial que no necesariamente es la completa u ptima, por lo tanto es necesario realizar varias iteraciones.
* Transformar el problema conceptual general en problemas particulares que tengan resolucin empleando Ciencia y Analtica de Datos.

#### 2a - Adquisicin de los datos (opcional)

* Obtencin de datos de alguna fuente externa o que sean de autora propia, fuente interna, con respecto a la organizacin con la que se est trabajando.
* Bsqueda de fuentes adicionales para completar la base de datos con la que se cuenta en caso de ser necesario.
* Proponer nuevos mtodos para recolectar datos de inters para la resolucin del problema.

#### 2 - Entendimiento de los datos

* Descripcin de los tipos de datos con los que se va a trabajar.
* Clasificacin de los datos (estructurados, no estructurados, semi-estructurados).
* Entender la correlacin que hay entre los datos con respecto a la problemtica que se busca resolver.

#### 3 - Preparacin de los datos

* Realizar limpieza de los datos, remover informacin que no sea relevante a la problemtica, control de las inconsistencias, estandarizacin de los mismos.
* En caso de tener datos faltantes, realizar la integracin con otro conjunto de datos.
* Generacin de las variables y conjuntos de datos con los que se va a trabaja.

> Evitar sesgos visualizando la informacin que se va a utilizar para evaluar el modelo.

#### 4 - Modelacin

* Seleccionar el modelo o tcnica que sera ms adecuado para la resolucin del problema.
* Aplicar tcnicas que sean consistentes con los datos que se tienen, los recursos y las necesidades del proyecto.
* Definir las mtricas de evaluacin y desempeo para evaluar calidad y fiabilidad del modelo.

> Se puede regresar a la fase anterior en caso de tener que retrabajar los datos y adaptarlos a las necesidades del modelo.

#### 5 - Evaluacin / Interpretacin

* Identificar y evaluar los resultados rigurosamente para tener confidencia de que son vlidos y el modelo repetible.
* Esperamos tener confidencia de que el modelo y los patrones que se extrajeron de los datos no presentan anormalidades, su visualizacuin y anlisis tienen correspondencia con los objetivos del proyecto, son de calidad y fiables.
* Tener una interpretacin de los datos que sea entendible para todos los **stakeholders** del proyecto y darles confidencia de que el modelo cumple las expectivas del proyecto

> Construir ambientes de pruebe que puedan simular condiciones de produccin.

#### 6 - Despliegue del Modelo / Resultados

* Mostrar los resultados obtenidos al cliente o interesados en el proyecto.
* Al momento de presentar los datos, mostrar indicadores como el **ROI (Retorno de la Inversin)** para sustentar que los recursos invertidos (tiempo, equipo de cmputo, personal, invesin monetaria, etc.) generan una utilidad o beneficio al cliente o interesados en el proyecto.
* Al ser un proceso iterativo, la retroalimentacin abre nuevas posibilidades e ideas que pueden complementar nuevos ciclos que se quieran ejecutar.
