### Pasos de la Metodología

#### 1 - Entendimiento del negocio / Problema

* Identificación de los objetivos y necesidades del proyecto.
* Tener una formulación inicial que no necesariamente es la completa u óptima, por lo tanto es necesario realizar varias iteraciones.
* Transformar el problema conceptual general en problemas particulares que tengan resolución empleando Ciencia y Analítica de Datos.

#### 2a - Adquisición de los datos (opcional)

* Obtención de datos de alguna fuente externa o que sean de autoría propia, fuente interna, con respecto a la organización con la que se esté trabajando.
* Búsqueda de fuentes adicionales para completar la base de datos con la que se cuenta en caso de ser necesario.
* Proponer nuevos métodos para recolectar datos de interés para la resolución del problema.

#### 2 - Entendimiento de los datos

* Descripción de los tipos de datos con los que se va a trabajar.
* Clasificación de los datos (estructurados, no estructurados, semi-estructurados).
* Entender la correlación que hay entre los datos con respecto a la problemática que se busca resolver.

#### 3 - Preparación de los datos

* Realizar limpieza de los datos, remover información que no sea relevante a la problemática, control de las inconsistencias, estandarización de los mismos.
* En caso de tener datos faltantes, realizar la integración con otro conjunto de datos.
* Generación de las variables y conjuntos de datos con los que se va a trabaja.

> Evitar sesgos visualizando la información que se va a utilizar para evaluar el modelo.

#### 4 - Modelación

* Seleccionar el modelo o técnica que sería más adecuado para la resolución del problema.
* Aplicar técnicas que sean consistentes con los datos que se tienen, los recursos y las necesidades del proyecto.
* Definir las métricas de evaluación y desempeño para evaluar calidad y fiabilidad del modelo.

> Se puede regresar a la fase anterior en caso de tener que retrabajar los datos y adaptarlos a las necesidades del modelo.

#### 5 - Evaluación / Interpretación

* Identificar y evaluar los resultados rigurosamente para tener confidencia de que son válidos y el modelo repetible.
* Esperamos tener confidencia de que el modelo y los patrones que se extrajeron de los datos no presentan anormalidades, su visualizacuión y análisis tienen correspondencia con los objetivos del proyecto, son de calidad y fiables.
* Tener una interpretación de los datos que sea entendible para todos los **stakeholders** del proyecto y darles confidencia de que el modelo cumple las expectivas del proyecto

> Construir ambientes de pruebe que puedan simular condiciones de producción.

#### 6 - Despliegue del Modelo / Resultados

* Mostrar los resultados obtenidos al cliente o interesados en el proyecto.
* Al momento de presentar los datos, mostrar indicadores como el **ROI (Retorno de la Inversión)** para sustentar que los recursos invertidos (tiempo, equipo de cómputo, personal, invesión monetaria, etc.) generan una utilidad o beneficio al cliente o interesados en el proyecto.
* Al ser un proceso iterativo, la retroalimentación abre nuevas posibilidades e ideas que pueden complementar nuevos ciclos que se quieran ejecutar.
