### Descripción del conjunto de datos. ¿Qué datos contiene?

Los datos continen todos los "posts" y comentarios publicados en la página de Reddit relacionados con el **cambio climático** donde ser mencionen los términos **"climate"** y **"change"** (clima y cambio) hasta fecha de 2022-09-01.

### ¿Cuáles son algunas de las preguntas de investigación que se podrían responder con estos datos? 


- ¿Qué opinión se tiene ante la problemática (análisis de sentimiento)?
- ¿Qué tendencia ha tenido la percepción de las personas con respecto al cambio climático? ¿Ha ido en incremento o decremento su interés o preocupación a lo largo de los años?
- ¿Existe algún patrón estacional con respecto a la percepción del cambio climático? (Por ejemplo hay determinadas épocas del año donde la preocupación sea menor o mayor?)
- ¿Qué desinformación se puede identificar sobre el tema que pueda ser dañina?
- ¿A través de estos datos se pueden identificar acciones, grupos de personas y organizaciones interesadas en combatir el cambio climático?
- ¿A partir del conjunto de datos se podría indentificar qué causas identifican las personas como las principales que contribuyen al cambio climático?
- Relacionando con el punto anterior, a partir del conjunto de datos, ¿se puede identificar cuáles son las principales acciones que las personas identifican para combatir el cambio climático? ¿Qué conocimiento se tiene sobre fuentes alternas o renovables de energía?
- ¿Qué opinión tienen las personas sobre la eficacia de los marcos regulatorios y políticas públicas y privadas para combatir el cambio climático?
- ¿Qué percepción se tienen sobre las empresas que contribuyen a la contaminación y las acciones que hacen para combatir el cambio climático?
- A partir de los datos análizados, ¿se podrían identificar acciones de impacto que ayuden por medio de los medios sociales a generar mayor conciencia y acciones de cambio en la población?
